# <b> Seats Won By Each Party

```python
total_seats_by_party = df.groupby('Party')['Won'].sum().reset_index().sort_values(by='Won', ascending=False)
print("\nTotal Seats Won by Each Party:")
print(total_seats_by_party.to_string(index=False))
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=total_seats_by_party, x='Party', y='Won', palette='bright')
plt.xticks(rotation=90)
plt.xlabel('Party')
plt.ylabel('Seats Won')
plt.title('Total Seats Won by Each Party')
plt.savefig('total_seats_won_by_party.png')
plt.show()
```

# <b> Total Seats Won in Each State by Each Party

```python
total_seats_by_state_party = df.groupby(['State', 'Party'])['Won'].sum().reset_index().sort_values(by=['State', 'Won'], ascending=[True, False])

current_state = None
for index, row in total_seats_by_state_party.iterrows():
    if row['State'] != current_state:
        current_state = row['State']
        state_name = current_state
        print(f"\nTotal Seats Won in {state_name} by Each Party:")
    print(f"{row['Party']}: {row['Won']}")
```

```python
plt.figure(figsize=(14, 10))
sns.barplot(x='Won', y='State', hue='Party', data=total_seats_by_state_party, dodge=False)
plt.title('Total Seats Won in Each State by Each Party')
plt.xlabel('Seats Won')
plt.ylabel('State')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
```

# <b> Total Seats in Each State

```python
total_seats_by_state = df.groupby('State')['Won'].sum().reset_index().sort_values(by='Won', ascending=False)
print("\nTotal Seats in Each State:")
print(total_seats_by_state.to_string(index=False))
```

```python
plt.figure(figsize=(18,10))
sns.barplot(data=total_seats_by_state, y='State', x='Won', palette='bright')
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Total Seats')
plt.title('Total Seats in Each State')
plt.show()
```
